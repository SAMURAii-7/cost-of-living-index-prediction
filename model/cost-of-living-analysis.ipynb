```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import init_notebook_mode, iplot
import pycountry
```

```python
df = pd.read_csv('./Cost_of_Living_Index_2022.csv',header=[1])
print(df.head())
```

# What are the most and least expensive contries to live?

## The most expensive:

```python
ms_exp_df = df.sort_values(by='Cost of Living Index').tail(15)
sorted_ms_exp_df = ms_exp_df.sort_values(by='Cost of Living Index', ascending=False)

sns.set_theme()
plt.figure(figsize=(16,8))
most_expbar = sns.barplot(x = 'Cost of Living Index', y = 'Country' , data = sorted_ms_exp_df, palette="rocket")

most_expbar.set_title('Top 15 Highest Cost of Living Index by Country')
```

## The least expensive:

```python
least_exp_df = df.sort_values(by='Cost of Living Index').head(15)

plt.figure(figsize=(16,8))
most_expbar = sns.barplot(x = 'Cost of Living Index', y = 'Country' , data = least_exp_df, palette="mako")

most_expbar.set_title('Top 15 Least Cost of Living by Country')
```

# The Cost of Living Index Map

```python
count = [ dict(
        type = 'choropleth',
        locations = df['Country'],
        locationmode='country names',
        z = df['Cost of Living Index'],
        text = df['Country'],
        colorscale = 'Viridis',
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick =False,
            title = 'Cost of Living Index'),
      ) ]
layout = dict(
    title = 'Cost of Living across the Globe',
    geo = dict(
        showframe = True,
        showcoastlines = True,
        projection = dict(
            type = 'Mercator'
        )
    )
)
fig = dict( data=count, layout=layout )
iplot( fig, validate=False, filename='d3-world-map' )
```
